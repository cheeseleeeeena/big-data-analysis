```python
df = pd.read_csv('articles_preprocessed.csv', sep='|')
df.head(6)
```

```python
horoscopes = ['水瓶座', '雙魚座', '牡羊座', '金牛座','雙子座','巨蟹座','獅子座','處女座','天秤座','天蠍座','射手座','摩羯座']
years = ['2021','2022']
months = ['1月','2月','3月','4月','5月','6月','7月','8月','9月','10月','11月','12月']

ignore_words = ['這樣','水瓶座', '雙魚座', '牡羊座', '金牛座','雙子座','巨蟹座','獅子座','處女座','天秤座','天蠍座','射手座','摩羯座', '可能', '已經', '可以', '能夠', '或許', '方面', '還是','不要','一直','再度','沒有','看到','做出','帶來','似乎','時候','也許','正在','進行','水瓶', '雙魚', '牡羊', '金牛','雙子','巨蟹','獅子','處女座','天秤','天蠍','射手','摩羯']

allowedPOS=['A', 'Na', 'Nb', 'Nc', 'Nv', 'VA', 'VAC', 'VB', 'VC', 'VH', 'VJ', 'VF', 'VE', 'D']


horoscopes_topkey = {}
for horoscope in horoscopes:

    yearly_topkey = {}
    for year in years:

        monthly_topkey = {}

        df_by_horoscope_yearly = df[(df.horoscope == horoscope) & (df.article_year == int(year))].sort_values("article_month", ascending=True)

        month_count = 0

        counter_by_year = Counter()

        for row in df_by_horoscope_yearly.token_pos:

            monthly_words = []
            for (word, pos) in eval(row):
                if (len(word) >= 2) & (pos in allowedPOS) & (word not in ignore_words):
                    monthly_words.append(word)

            counter_by_month = Counter(monthly_words)
            counter_by_year += counter_by_month
            
            monthly_topkey[months[month_count]] = counter_by_month.most_common(100)

            if month_count < 11:
                month_count += 1

        monthly_topkey['全年'] = counter_by_year.most_common(100)

        yearly_topkey[year] = monthly_topkey

    horoscopes_topkey[horoscope] = yearly_topkey

      
```

```python
df_horoscopes_topkey = pd.DataFrame(list(horoscopes_topkey.items()), columns = ['horoscope','top_keys'])

df_horoscopes_topkey.to_csv('horoscope_topkey_by_month.csv', sep='|', index=False)
```
